## Exploratory Data Analysis

As we only have 150 rows, no missing values, and no outliers, there is no data cleaning to perform.

```python
sns.pairplot(df, hue='species')
plt.savefig("./images/pairplot.png")
```

> There is quite  bit of interesting information in these plots. We can see some clear, distinct areas in the scatter plots for each type of species. Same for the histograms with the different species dominating some areas compared to the other species.
>
> Let's take a deeper dive into the plots. 

### Petal Length

```python
sns.histplot(data=df, x='petal_length', hue='species', bins=20, kde=True)
plt.title("Distribution of Petal Length (cm)")
plt.xlabel("Petal Length (cm)")
plt.savefig('./images/petal_length_hist.png')
```

> In this histogram, we can clearly see that Setosas' petal length only ranges from 1-2 cm, and there is a considerable gap between Setosas and the other 2 species, Versicolor, which ranges from 3-5.5 cm, and Virginica, which ranges from 4-7 cm.
>
> Versicolor and Virginica have some overlap, but Setosa is in its own area. 

#### Petal Length vs Petal Width

```python
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title("Petal Length vs Petal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)");
plt.savefig('./images/petal_length_vs_petal_width.png')
```

#### Petal Length vs Sepal Length

```python
sns.scatterplot(data=df, x='petal_length', y='sepal_length', hue='species')
plt.title("Petal Length vs Sepal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Sepal Length (cm)");
plt.savefig('./images/petal_length_vs_sepal_length.png')
```

#### Petal Length vs Sepal Width

```python
sns.scatterplot(data=df, x='petal_length', y="sepal_width", hue='species')
plt.title("Petal Length vs Sepal Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Sepal Width (cm)");
plt.savefig('./images/petal_length_vs_sepal_width.png')
```

### Sepal Length

```python
sns.histplot(data=df, x='sepal_length', hue='species', kde=True)
plt.title("Distribution of Sepal Length (cm)")
plt.xlabel('Sepal Length (cm)')
plt.savefig('./images/sepal_length_hist.png')
```

> Unlike with Petal Length, all 3 species overlap in this histogram. The Setosa range for sepal length is from ~4.5-6 cm, Versicolor from ~4.5-7.5 cm, and Virginica ~4.5-8 cm. While they all have the same length on the low side, their higher ranges are different, so the distribution for each overlap, but they all have different peaks and different widths, means they have different means and standard deviations.

### Petal Width

```python
sns.histplot(data=df, x='petal_width', hue='species', kde=True)
plt.title("Distribution of Petal Width (cm)")
plt.xlabel('Petal Width (cm)')
plt.savefig('./images/petal_width_hist.png')
```

> Similar to Petal Length, there are more obvious separations between the different species. The Septosa range goes from 0-0.5 approximately. There is a decent gap between Setosa and the other 2 speciers. Versicolor have a range from around 0.75-2, and Virginica goes from 1-2.5, causing an overlap.

### Sepal Width

```python
sns.histplot(data=df, x='sepal_width', hue='species', kde=True)
plt.title("Distribution of Sepal Width (cm)")
plt.xlabel('Sepal Width (cm)')
plt.savefig('./images/sepal_width_hist.png')
```
